```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("spam.csv",encoding="latin-1")
```

```python
df=data.copy()
df.head()
```

```python
df1=df.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1)
```

```python
df1.head()
```

```python
# rename the column
df1=df1[["v2","v1"]].rename(columns={"v2":"email","v1":"label"})
df1.head()
```

```python
# rename the target variable
df1["label"]=df1["label"].map({"ham":0,"spam":1})
df1["label"].value_counts()
```

```python
df1.isnull().sum()
```

```python
# dependent & independent variable
```

```python
X=df1["email"]
Y=df1["label"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10,stratify=Y)
```

```python
X_train.tail()
```

```python
# creation of bag of words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer=CountVectorizer()
new_X_train=vectorizer.fit(X_train)
```

```python
new_X_train
```

```python
X_train_vector=vectorizer.transform(X_train)
```

```python
X_train_vector
```

```python
X_test_vector=vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB,GaussianNB
```

```python
mnb=MultinomialNB(alpha=0.01)
mnb.fit(X_train_vector,Y_train)
```

```python
predict_test=mnb.predict(X_test_vector)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test,predict_test)
```

```python
confusion_matrix(Y_test,predict_test)
```

```python
print(classification_report(Y_test,predict_test))
```

```python
def enter_text(message):
    if mnb.predict(vectorizer.transform([message]))==1:
        print("This a  spam")
    else:
        print("Normal message")
```
